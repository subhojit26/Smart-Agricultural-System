```python
import re
import numpy as np
from typing import Dict, Union

class RicePesticideAnalyzer:
    def __init__(self):
        # Define rice-compatible ingredients with safe and optimal ranges
        self.rice_compatible_ingredients = {
            'azoxystrobin': {'safe_range': (0.1, 15.0), 'optimal_range': (5.0, 12.0)},
            'propiconazole': {'safe_range': (0.1, 10.0), 'optimal_range': (3.0, 7.0)},
            'tebuconazole': {'safe_range': (0.1, 7.0), 'optimal_range': (2.0, 5.0)},
            'pyraclostrobin': {'safe_range': (0.1, 6.0), 'optimal_range': (1.5, 4.5)},
            'difenoconazole': {'safe_range': (0.1, 5.0), 'optimal_range': (1.0, 3.5)},
            'copper oxychloride': {'safe_range': (0.1, 5.0), 'optimal_range': (1.0, 4.0)}
        }

        # Harmful ingredients to avoid
        self.harmful_ingredients = ['arsenic', 'mercury', 'lead']

        # Optimal application conditions for rice
        self.optimal_conditions = {
            'ph_range': (6.5, 7.5),
            'optimal_nutrients': {
                'n': (10.0, 15.0),
                'p': (10.0, 15.0),
                'k': (10.0, 15.0)
            }
        }

    def parse_ocr_text(self, ocr_text: str) -> Dict[str, Union[float, Dict]]:
        """
        Parse OCR text and extract ingredient information
        """
        # Normalize and clean the text
        ocr_text = ocr_text.lower()

        # Initialize parsed data dictionary
        parsed_data = {
            'ingredients': {},
            'nutrients': {},
            'ph': None
        }

        # Extract ingredients
        ingredient_pattern = r'(\w+(?:\s+\w+)*)\s+(\d+\.?\d*)\s*g'
        ingredient_matches = re.findall(ingredient_pattern, ocr_text)

        for ingredient, quantity in ingredient_matches:
            quantity = float(quantity)

            # Normalize ingredient names
            normalized_ingredient = self._normalize_ingredient_name(ingredient)

            if normalized_ingredient in self.rice_compatible_ingredients:
                parsed_data['ingredients'][normalized_ingredient] = quantity

            # Check for nutrients
            nutrient_mapping = {
                'nitrogen': 'n',
                'phosphate': 'p',
                'potash': 'k',
                'magnesium': 'mg',
                'sulfur': 's'
            }

            for nutrient_name, nutrient_code in nutrient_mapping.items():
                if nutrient_name in ingredient:
                    parsed_data['nutrients'][nutrient_code] = quantity

        # Extract pH
        ph_match = re.search(r'ph\s*[Vv]alue\s*(\d+\.?\d*)-?(\d+\.?\d*)?', ocr_text)
        if ph_match:
            parsed_data['ph'] = float(ph_match.group(1))

        return parsed_data

    def _normalize_ingredient_name(self, ingredient: str) -> str:
        """
        Normalize ingredient names for consistent matching
        """
        normalization_map = {
            'azoxystrobin': 'azoxystrobin',
            'azo xy st robin': 'azoxystrobin',
            'propiconazole': 'propiconazole',
            'prop i con az ole': 'propiconazole',
            'tebuconazole': 'tebuconazole',
            'te bu co nazo le': 'tebuconazole',
            'pyraclostrobin': 'pyraclostrobin',
            'pyr a clost robin': 'pyraclostrobin',
            'difenoconazole': 'difenoconazole',
            'd if eno co nazo le': 'difenoconazole',
            'copper oxychloride': 'copper oxychloride'
        }

        return normalization_map.get(ingredient, ingredient)

    def analyze_compatibility(self, ocr_text: str) -> Dict:
        """
        Comprehensive analysis of pesticide compatibility for rice crops
        """
        # Parse OCR text
        parsed_data = self.parse_ocr_text(ocr_text)

        # Calculate individual compatibility scores
        scores = {
            'active_ingredients': self._analyze_active_ingredients(parsed_data['ingredients']),
            'nutrient_balance': self._analyze_nutrients(parsed_data['nutrients']),
            'ph_compatibility': self._analyze_ph(parsed_data['ph'])
        }

        # Calculate overall compatibility score
        overall_score = (
            scores['active_ingredients'] * 0.5 +
            scores['nutrient_balance'] * 0.3 +
            scores['ph_compatibility'] * 0.2
        )

        # Generate recommendation
        recommendation = self._generate_recommendation(overall_score)

        return {
            'parsed_data': parsed_data,
            'compatibility_scores': scores,
            'overall_compatibility_score': round(overall_score * 100, 2),
            'recommendation': recommendation
        }

    def _analyze_active_ingredients(self, ingredients: Dict[str, float]) -> float:
        """
        Analyze compatibility of active ingredients
        """
        if not ingredients:
            return 0.0

        ingredient_scores = []
        for ingredient, quantity in ingredients.items():
            if ingredient in self.rice_compatible_ingredients:
                safe_range = self.rice_compatible_ingredients[ingredient]['safe_range']
                optimal_range = self.rice_compatible_ingredients[ingredient]['optimal_range']

                if safe_range[0] <= quantity <= safe_range[1]:
                    # Score based on proximity to optimal range
                    if optimal_range[0] <= quantity <= optimal_range[1]:
                        ingredient_scores.append(1.0)  # Optimal
                    else:
                        # Partial score based on distance from optimal range
                        distance = min(
                            abs(quantity - optimal_range[0]),
                            abs(quantity - optimal_range[1])
                        )
                        ingredient_scores.append(max(0.5, 1 - (distance / 10)))
                else:
                    ingredient_scores.append(0.2)  # Outside safe range

        return np.mean(ingredient_scores) if ingredient_scores else 0.0

    def _analyze_nutrients(self, nutrients: Dict[str, float]) -> float:
        """
        Analyze nutrient balance
        """
        if not nutrients:
            return 0.5  # Neutral score if no nutrients

        nutrient_scores = []
        for nutrient, value in nutrients.items():
            if nutrient in self.optimal_conditions['optimal_nutrients']:
                optimal_range = self.optimal_conditions['optimal_nutrients'][nutrient]
                if optimal_range[0] <= value <= optimal_range[1]:
                    nutrient_scores.append(1.0)
                else:
                    # Partial score based on deviation
                    deviation = abs(value - np.mean(optimal_range))
                    nutrient_scores.append(max(0, 1 - (deviation / 10)))

        return np.mean(nutrient_scores) if nutrient_scores else 0.5

    def _analyze_ph(self, ph: Union[float, None]) -> float:
        """
        Analyze pH compatibility
        """
        if ph is None:
            return 0.5  # Neutral score if pH unknown

        optimal_range = self.optimal_conditions['ph_range']
        if optimal_range[0] <= ph <= optimal_range[1]:
            return 1.0

        # Partial score based on deviation from optimal range
        deviation = min(
            abs(ph - optimal_range[0]),
            abs(ph - optimal_range[1])
        )
        return max(0, 1 - (deviation / 2))

    def _generate_recommendation(self, compatibility_score: float) -> str:
        """
        Generate recommendation based on compatibility score
        """
        if compatibility_score >= 0.8:
            return "Highly Recommended - Excellent for Rice Crops"
        elif compatibility_score >= 0.6:
            return "Conditionally Recommended - Apply with Caution"
        elif compatibility_score >= 0.4:
            return "Limited Compatibility - Careful Consideration Needed"
        else:
            return "Not Recommended for Rice Crops"

def main():
    # Example OCR text
    ocr_text = """
    CROP GUARD ULTRA PLUS COMPOSITION INFORMATION
    Serving Size: 30 g
    Ingredients Per 100 g Per 30 g Serve
    A zo xy st robin 12.0 g 3.6 g
    Prop i con az ole 10.0 g 3.0 g
    Te bu co nazo le 7.0 g 2.1 g
    Pyr a clost robin 6.0 g 1.8 g
    D if eno co nazo le 5.0 g 1.5 g
    Copper Oxy chloride 4.0 g 1.2 g
    NUTRIENT INFORMATION
    Total Nitrogen( N) 12.0 g 3.6 g
    Phosphate( PO) 15.0 g 4.5 g
    Potash( KO) 15.0 g 4.5 g
    Magnesium( Mg) 2.0 g 0.6 g
    Sulfur( S) 2.0 g 0.6 g
    PROPERTIES
    pH Value 6.8-7.4
    Density 1.15 g/ ml
    Store in cool, dry place. Keep container tightly closed.
    """

    # Create analyzer instance
    analyzer = RicePesticideAnalyzer()

    # Analyze compatibility
    result = analyzer.analyze_compatibility(ocr_text)

    # Display results
    print("Pesticide Compatibility Analysis for Rice Crops")
    print("=" * 50)
    print(f"Overall Compatibility Score: {result['overall_compatibility_score']}%")
    print("\nCompatibility Breakdown:")
    for category, score in result['compatibility_scores'].items():
        print(f"{category.replace('_', ' ').title()}: {score * 100:.2f}%")

    print("\nRecommendation:")
    print(result['recommendation'])

    print("\nParsed Data:")
    print("Ingredients:", result['parsed_data']['ingredients'])
    print("Nutrients:", result['parsed_data']['nutrients'])
    print("pH:", result['parsed_data']['ph'])

if __name__ == "__main__":
    main()
```
